```python
import numpy as np
us = [0.1, 0.2, 0.35, 0.5, 0.65]
alpha = 0.7
x = []
y = []
for u in us:
    path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_u' + str(u)
    df = pd.read_csv(path)
    infected = []
    susceptible = []
    time = []
    is_edges = []
    tot_edges = []
    for inf in df['infected'] :
        infected.append(inf/300)
        x.append(inf/300)

    for sus in df['susceptible'] :
        susceptible.append(sus/300)

    for ss in df['[SS]']:
        tot_edges.append(ss)

    for i in range(0,len(tot_edges)):
        tot_edges[i] += df['[II]'][i]
        tot_edges[i] += df['[IS]'][i]
        is_edges.append(df['[IS]'][i]/tot_edges[i])
        y.append(df['[IS]'][i]/tot_edges[i])

    plt.scatter(infected, is_edges, label='u = ' + str(u))
    plt.xlabel('Fraction of Nodes Infected') 
    plt.ylabel('Fraction of Edges Disjunct') 
    plt.title('Proportion of [IS] vs. Proportion of Infecteds: Barabasi-Albert')
c1, b1, a1 = np.polyfit(x, y, deg=2)
xseq = np.linspace(0, 1, num=100)
# Plot regression line
plt.plot(xseq, a1 + b1 * xseq + (c1 * xseq)*xseq, color="k", lw=2.5)
plt.legend()
```
