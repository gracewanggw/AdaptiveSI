```python
alphas = [0.5, 0.65, 0.7, 0.75, 0.8, 0.85, 0.87, 0.9, 0.95]

for alpha in alphas:
    infected = []
    susceptible = []
    time = []
    is_edges = []
    path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_ER'
    df = pd.read_csv(path)
    for idx in df.index:
        infected.append(df['infected'][idx]/300)
        susceptible.append(df['susceptible'][idx]/300)
        time.append(df['t'][idx])
        is_edges.append(df['[IS]'][idx]/1500)
    
    plt.plot(infected, is_edges, label='alpha = ' + str(alpha))
    plt.xlabel('Fraction of Nodes Infected') 
    plt.ylabel('Fraction of Edges Disjunct') 
    plt.legend()

```
