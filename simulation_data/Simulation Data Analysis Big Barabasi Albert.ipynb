```python
alphas = [0.5, 0.65, 0.7, 0.75, 0.8, 0.85, 0.87, 0.9, 0.95]

for alpha in alphas:
    path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_big'
    df = pd.read_csv(path)
    infected = []
    susceptible = []
    time = []
    is_edges = []
    tot_edges = []
    for inf in df['infected'] :
        infected.append(inf/300)

    for sus in df['susceptible'] :
        susceptible.append(sus/300)

    for ss in df['[SS]']:
        tot_edges.append(ss)

    for i in range(0,len(tot_edges)):
        tot_edges[i] += df['[II]'][i]
        tot_edges[i] += df['[IS]'][i]
        is_edges.append(df['[IS]'][i]/tot_edges[i])
    
    plt.plot(infected, is_edges, label='alpha = ' + str(alpha))
    plt.xlabel('Fraction of Nodes Infected') 
    plt.ylabel('Fraction of Edges Disjunct') 
    plt.title('Proportion of [IS] vs. Proportion of Infecteds: Barabasi-Albert')
    plt.legend()
```
