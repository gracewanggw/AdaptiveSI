# Plot Fraction of edges and nodes

```python
alpha = 0.5
infected = []
susceptible = []
time = []
for i in range (1, 10):
    path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(i)
    df = pd.read_csv(path)
    
    for idx in df.index:
        infected.append(df['infected'][idx])
        susceptible.append(df['susceptible'][idx])
        time.append(df['t'][idx])

plt.plot(time, infected, 'o', color='green')
```
