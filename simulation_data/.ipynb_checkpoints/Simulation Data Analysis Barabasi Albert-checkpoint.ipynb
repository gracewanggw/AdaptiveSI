```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
alphas = [0.5, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95]
colors = ['green', 'blue', 'brown', 'cyan', 'purple', 'pink', 'olive', 'gray']
```

```python
def get_mean_data(alpha):
    res_inf = []
    res_sus = []
    ts = []
    vals_inf = {} #t: [n1, n2, ...]
    vals_sus = {}
    
    path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(1)
    df = pd.read_csv(path)
    t=0
    for inf in df['infected']:
        vals_inf[t] = [inf]
        ts.append(t)
        t += 1
    t=0
    for sus in df['susceptible']:
        vals_sus[t] = [sus]
        t += 1
    
    for i in range(2, 10):
        path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(i)
        df = pd.read_csv(path)
        
        t=0
        for inf in df['infected']:
            if vals_inf[t]:
                vals_inf[t].append(inf)
            else:
                vals_inf[t] = [inf]
            t += 1
        t=0
        for sus in df['susceptible']:
            vals_sus[t].append(sus)
            t += 1
    
    for t in vals_inf:
        res_inf.append(sum(vals_inf[t])/len(vals_inf[t]))

    for t in vals_sus:
        res_sus.append(sum(vals_sus[t])/len(vals_sus[t]))
    
    return res_inf, res_sus, ts
        
    
```

# Nodes Infected/ Susceptible Over Time

```python
alpha = 0.5
path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(1)
df = pd.read_csv(path)
    
plt.plot(df['t'], df['infected'], color='red', label='infected')
plt.plot(df['t'], df['susceptible'], color='blue', label='susceptible')
plt.xlabel('t') 
plt.ylabel('N(t)') 
plt.title("Alpha = " + str(alpha))
plt.legend()
    
for i in range (2, 10):
    path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(i)
    df = pd.read_csv(path)
    
    plt.plot(df['t'], df['infected'], color='red')
    plt.plot(df['t'], df['susceptible'], color='blue')


mean_inf, mean_sus, ts = get_mean_data(alpha)
plt.plot(ts, mean_inf, color='purple')
plt.plot(ts, mean_sus, color='cyan')
```

# Plot Fraction of edges and nodes

```python
alpha = 0.5
infected = []
susceptible = []
time = []
for i in range (1, 10):
    path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(i)
    df = pd.read_csv(path)
    
    for idx in df.index:
        infected.append(df['infected'][idx])
        susceptible.append(df['susceptible'][idx])
        time.append(df['t'][idx])

plt.plot(time, infected, 'o', color='green')
```
