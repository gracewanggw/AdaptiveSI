# Nodes Infected/ Susceptible Over Time

```python
def get_mean_data_nodes(alpha):
    res_inf = []
    res_sus = []
    ts = []
    vals_inf = {} #t: [n1, n2, ...]
    vals_sus = {}
    
    path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(1)
    df = pd.read_csv(path)
    t=0
    for inf in df['infected']:
        vals_inf[t] = [inf]
        ts.append(t)
        t += 1
    t=0
    for sus in df['susceptible']:
        vals_sus[t] = [sus]
        t += 1
    
    for i in range(2, 10):
        path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(i)
        df = pd.read_csv(path)
        
        t=0
        for inf in df['infected']:
            if t in vals_inf:
                vals_inf[t].append(inf)
            else:
                vals_inf[t] = [inf]
                ts.append(t)
            t += 1
        t=0
        for sus in df['susceptible']:
            if t in vals_sus:
                vals_sus[t].append(sus)
            else:
                vals_sus[t] = [sus]
            t += 1
    
    for t in vals_inf:
        res_inf.append(sum(vals_inf[t])/len(vals_inf[t]))

    for t in vals_sus:
        res_sus.append(sum(vals_sus[t])/len(vals_sus[t]))
    
    return res_inf, res_sus, ts
        
    
```

```python
alpha = 0.5
path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(1)
df = pd.read_csv(path)
    
plt.plot(df['t'], df['infected'], color='red', label='infected')
plt.plot(df['t'], df['susceptible'], color='blue', label='susceptible')
    
for i in range (2, 10):
    path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(i)
    df = pd.read_csv(path)
    
    plt.plot(df['t'], df['infected'], color='red')
    plt.plot(df['t'], df['susceptible'], color='blue')


mean_inf, mean_sus, ts = get_mean_data_nodes(alpha)
plt.plot(ts, mean_inf, color='maroon', label='avg. infected')
plt.plot(ts, mean_sus, color='cyan', label='avg. susceptible')
plt.xlabel('t') 
plt.ylabel('# of Nodes') 
plt.title("Alpha = " + str(alpha))
plt.legend()
```

# Edges Over Time

```python
def get_mean_data_edges(alpha):
    res_ii = []
    res_ss = []
    res_is = []
    
    ts = []
    
    vals_ii = {} #t: [n1, n2, ...]
    vals_ss = {}
    vals_is = {}
    
    path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(1)
    df = pd.read_csv(path)
    t=0
    for inf in df['[II]']:
        vals_ii[t] = [inf]
        ts.append(t)
        t += 1
    t=0
    for sus in df['[SS]']:
        vals_ss[t] = [sus]
        t += 1
    t=0
    for mix in df['[IS]']:
        vals_is[t] = [mix]
        t += 1
    
    for i in range(2, 10):
        path = '/Users/gracewang/Desktop/Dartmouth/MATH36/AdaptiveSIS/simulation_data/alpha' + str(alpha) + '/simulation_data_' + str(alpha) + '_trial' + str(i)
        df = pd.read_csv(path)
        
        t=0
        for inf in df['[II]']:
            if t in vals_ii:
                vals_ii[t].append(inf)
            else:
                vals_ii[t] = [inf]
                ts.append(t)
            t += 1
        t=0
        for sus in df['[SS]']:
            if t in vals_ss:
                vals_ss[t].append(sus)
            else:
                vals_ss[t] = [sus]
            t += 1
        t=0
        for mix in df['[IS]']:
            if t in vals_is:
                vals_is[t].append(mix)
            else:
                vals_is[t] = [mix]
            t += 1
    
    for t in vals_ii:
        res_ii.append(sum(vals_ii[t])/len(vals_ii[t]))

    for t in vals_ss:
        res_ss.append(sum(vals_ss[t])/len(vals_ss[t]))
    
    for t in vals_is:
        res_is.append(sum(vals_is[t])/len(vals_is[t]))
    
    return res_ii, res_ss, res_is, ts
        
```

# Plot Fraction of edges and nodes

```python
alphas = [0.5, 0.65, 0.7, 0.75, 0.8, 0.85, 0.87, 0.9, 0.95]

for alpha in alphas:
    infected = []
    susceptible = []
    time = []
    is_edges = []
    mean_ii, mean_ss, mean_is, ts = get_mean_data_edges(alpha)
    mean_inf, mean_sus, ts = get_mean_data_nodes(alpha)
    for t in ts :
        infected.append(mean_inf[t]/20)
        susceptible.append(mean_sus[t]/20)
        is_edges.append(mean_is[t]/75)
    
    plt.plot(infected, is_edges, label='alpha = ' + str(alpha))
    plt.xlabel('Fraction of Nodes Infected') 
    plt.ylabel('Fraction of Edges Disjunct') 
    plt.title('Proportion of [IS] vs. Proportion of Infecteds: Barabasi-Albert')
    plt.legend()
```
